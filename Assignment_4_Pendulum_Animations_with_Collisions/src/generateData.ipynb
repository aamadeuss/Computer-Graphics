```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
times = np.arange(0,5000,10)
sines = np.sin(times*2*np.pi/np.max(times))
plt.plot(times,sines)

new_df = pd.DataFrame()
new_df["time"]=times
new_df["testpilot"]=sines
```

```python
new_df.to_csv("testpilot.csv",index=False)
```
